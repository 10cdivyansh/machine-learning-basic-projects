```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url=r"C:\Users\Divyansh\Desktop\udemy machine learning\codes\polynomial regression\positional salary\Position_Salaries.csv"
dataset=pd.read_csv(url)
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(dataset)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly =poly_reg.fit_transform(X)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
X_poly
```

```python
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color="red")
plt.plot(X,lin_reg.predict(X),color="blue")
plt.plot(X_grid,lin_reg2.predict(poly_reg.fit_transform(X_grid)),color="black")
plt.title("Linear Regression and polynomial regression")
plt.xlabel("Position lavel")
plt.ylabel("Salary")
plt.show()
```

```python
#for linear regression
lin_reg.predict([[6.5]])
```

```python
#for polynomial regression
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
